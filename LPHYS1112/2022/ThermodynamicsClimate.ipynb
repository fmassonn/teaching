```python
# Libraries, imports, etc.
import numpy             as np
import matplotlib.pyplot as plt

# Physical constants
h     = 6.62607015e-34 # [m 2 / kg  /   s]  Planck's constant
sigma = 5.670374419e-8 # [W   / m^2 / K^4]  Stefan-Boltzmann constant
c     = 299792458      # [m / s]            Speed of light
kB    = 1.380649e-23   # [J / K]            Boltzmann constant

# Functions
def PlanckLaw(nu, T):
    return 2 * h * nu ** 3 / c ** 2 * 1 / (np.exp(h * nu / kB * T) - 1)

# Plot blackbody emission curve
T  = 5780                                   # Emission temperature
waveLength = np.linspace(0.0, 6000 * 1e-9)  # [m] wavelength as a variable 
                                            # (up to 6000 nm)
frequency  = c / waveLength                 # equivalent frequency 
spectrum = PlanckLaw(frequency, T)

fig, ax = plt.subplots(1, 1, dpi = 300)
ax.plot(nu, spectrum)
```
